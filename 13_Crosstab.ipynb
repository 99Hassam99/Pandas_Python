In statistics,
# Contingency tables also known as cross tabulation or crosstab is a type of table in a matrix format that displays the (multivariant) frequency distribution of variables.

```python
import pandas as pd
df = pd.read_excel('13_survey.xls')
df
```

```python
# crosstab is defined directly on package and not on a DATAFRAME and DF should be inside the package
pd.crosstab(df.Nationality,df.Handedness)
```

```python
pd.crosstab(df.Sex,df.Handedness)
```

```python
pd.crosstab(df.Age,df.Handedness,margins=True) # for total we define margins=True
```

```python
pd.crosstab(df.Sex,[df.Handedness,df.Nationality],margins=True)
```

```python
# for percentage etc we use normalize by passing index
```

```python
pd.crosstab(df.Sex,df.Handedness,normalize='index')
```

```python
# to find average age of males and females
```

```python
import numpy as np
pd.crosstab(df.Sex,df.Handedness,values=df.Age,aggfunc=np.average)
```
