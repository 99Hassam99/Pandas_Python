# Replace Function

```python
import pandas as pd
import numpy as np
df = pd.read_csv('06_nyc_data.csv')
df
```

```python
new_df = df.replace(-99999,np.nan)
new_df
```

```python
new_df=df.replace({
    'temperature':-99999,
    'windspeed':-99999,
    'event':'0' # replace values using dictionary
},np.nan)
new_df
```

```python
df=pd.read_csv('06_nyc_data.csv')
df
```

```python
new_df=df.replace({
    -99999: np.nan,
    'No Event':'Sunny'
}) # mapping
new_df
```

```python
df=pd.read_csv('06_nyc_data.csv')
df
```

# for this to replace different values like mph, F, C we use RegX(regular expression),which are used to detect pattern and whenever we want to replace patterns(mph c f) we use RegX

```python
new_df=df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex=True)
new_df
```

# Replacing list of values with another list of values.
#For this we will create a new DataFrame

```python
df = pd.DataFrame({
    'score':['exceptional','average','good','poor','average','exceptional'],
    'student':['rob','saya','asim','bajwa','munir','tom']
})
df
```

```python
new_df=df.replace(['poor','average','good','exceptional'],[1,2,3,4])
new_df
```
