# Reshaping DataFrame Using Melt
Melt is used to transform or reshape Data

```python
import pandas as pd
df = pd.read_csv('11_weather.csv')
df
```

```python
df1 = pd.melt(df, id_vars=['day'])
df1
```

```python
df1 = pd.melt(df, id_vars=['day'])
df1[df1['variable'] == 'chichago']
```

```python
df1 = pd.melt(df, id_vars=['day'])
df1[df1['variable'] == 'peshawar']
```

```python
df1 = pd.melt(df, id_vars=['day'])
df1[df1['variable'] == 'berlin']
```
