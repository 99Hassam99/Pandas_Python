# Pivot
Pivot allows to reshape or transforms the dataFrame

```python
import pandas as pd
df = pd.read_csv('10_ptdata.csv')
df
```

```python
df.pivot(index='date',columns='city')
```

```python
df.pivot(index='date',columns='city',values='humidity')
```

```python
df.pivot(index='humidity',columns='city')
```

```python
df.pivot(index='temp',columns='city')
```

```python
df.pivot(index='date',columns='city',values='temp')
```

# Pivot Table
Pivot table is used to summarize and aggregate data inside dataframe. It allows us to summarize and aggregate our tabular data.

```python
df = pd.read_csv('10_ptdata2.csv')
df
```

```python
df.pivot_table(index='city',columns='date')
```

```python
# it has already done an aggregate func but we can define it by our self as:
df.pivot_table(index='city',columns='date',aggfunc='sum')
```

```python
df.pivot_table(index='city',columns='date',aggfunc='count')
```

```python
df.pivot_table(index='city',columns='date',aggfunc='mean')
```

```python
df.pivot_table(index='city',columns='date',aggfunc='max')
```

```python
df.pivot_table(index='city',columns='date',aggfunc='min')
```

```python
df.pivot_table(index='city',columns='date',aggfunc='diff')
```

```python
df.pivot_table(index='city',columns='date',margins=True)
```

```python
df.pivot_table(index='city',columns='date',margins=False)
```
