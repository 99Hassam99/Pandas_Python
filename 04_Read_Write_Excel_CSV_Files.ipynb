```python
# if we see there are too much n.a. so for this we use converters
# lets define
def convert_People_cell(cell):
    if cell == "n.a.":
        return 'sam walton'
    return cell
def convert_eps_cell(cell):
    if cell == 'not available':
        return None
    return cell
df = pd.read_excel("Data2.xlsx","Sheet2",converters={
    'People':convert_People_cell,
    'eps':convert_eps_cell
})
df
```

```python
# it was all about reading excel file
# now let us cover writing to excel file
```

```python
df.to_excel("new_data1.xlsx",index=False, sheet_name="Stocks", startrow=1,startcol=2)
```

```python
# if we have 2 dataframes and we want to write them together to the same excel file in different sheets
# for this we have a class called excel writer
```

```python
# lets create
df_stocks = pd.DataFrame({
    'tickers':['Google','WMT','Sony'],
    'price':[845,43,121],
    'pe':[12.3,43.54,23.12],
    'eps':[27.43,5.2,2.12]
})

df_weather = pd.DataFrame({
    'day':['14/10/24','10/6/24','19/9/24'],
    'temperature':[20,45,33],
    'event':['rain','hot','sunny']
})

with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer,sheet_name="stocks")
    df_weather.to_excel(writer,sheet_name="weather")
```
